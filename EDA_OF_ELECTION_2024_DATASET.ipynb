```python
#Vote distrubation by cities

import pandas as pd
import numpy as np

# Generate synthetic data
np.random.seed(42)
num_cities = 10
cities = [f"City {i}" for i in range(1, num_cities + 1)]
votes = np.random.randint(1000, 5000, size=num_cities)

synthetic_df = pd.DataFrame({"City": cities, "Votes": votes})
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.pie(synthetic_df["Votes"], labels=synthetic_df["City"], autopct="%1.1f%%", startangle=90)
plt.title("Vote Distribution by City")
plt.axis("equal")  # Equal aspect ratio ensures the pie chart is drawn as a circle.
plt.show()
```

```python
#corelation between sex and age

# Convert 'sex' column to numerical values if it contains strings
if data['sex'].dtype == 'object':
    data['sex'] = data['sex'].astype('category').cat.codes

# Calculate correlation
correlation = data['sex'].corr(data['age'])
print("Correlation between sex and age:", correlation)
```

```python
gender_age_group = data.groupby("sex")["age"].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(gender_age_group.index, gender_age_group.values, color=["skyblue", "salmon"])
plt.xlabel("sex")
plt.ylabel("Average Age")
plt.title("Correlation between Gender and Average Age")
plt.ylim(20, 70)  # Adjust the y-axis limits
plt.xticks(rotation=0)
plt.show()
```

```python
#correlation between employment and income

#correlation between employment and income

# Convert 'income' column to numeric, handling non-numeric values
data['income'] = pd.to_numeric(data['income'], errors='coerce')

income_employ = data.groupby("employment")["income"].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(income_employ.index, income_employ.values, color="skyblue") # Fix: Use income_employ instead of income_by_status
plt.xlabel("Employment Status")
plt.ylabel("Average Income")
plt.title("Correlation between Employment Status and Income")
plt.xticks(rotation=0)
plt.show()
```
